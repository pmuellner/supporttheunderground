```python
import pandas as pd
import sqlalchemy
```

```python
beyms = pd.read_csv("data/beyms.csv")["user_id"].tolist()
ms = pd.read_csv("data/ms.csv")["user_id"].tolist()
```

```python
list_of_tuples = []

chunksize = 10**6
i = 0
for chunk in pd.read_csv("data/events.tsv", sep="\t", chunksize=chunksize):
    subchunk = chunk[chunk["user_id"].isin(beyms+ms)]
    data_tuples = [(user_id, artist_id, album_id, track_id, timestamp) for _, user_id, artist_id, album_id, track_id, timestamp in subchunk.itertuples()]
    list_of_tuples.extend(data_tuples)
    i += 1
    
    print(i)
```

```python
events_df = pd.DataFrame.from_records(list_of_tuples, columns=["user_id", "artist_id", "album_id", "track_id", "timestamp"])
events_df.head()
```

```python
events_df["user_id"].isin(beyms).sum()
```

```python
connection = sqlalchemy.create_engine("mysql+pymysql://root:1234@localhost:3306/music_recommender_db")
df = pd.read_sql(con=connection, sql="SELECT * FROM events WHERE user_id IN " + str(tuple(beyms)))
df.head()
```
